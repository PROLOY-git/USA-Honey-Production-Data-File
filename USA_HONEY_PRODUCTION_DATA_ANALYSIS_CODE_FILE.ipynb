# Q2. Which are the top 5 Honey producing state?

```python
df
```

```python
df.groupby('state').sum()["production"]
```

```python
df.groupby('state').sum()["production"].reset_index()
```

```python
new_df=df.groupby("state").sum()['production'].reset_index()
```

```python
new_df
```

```python
new_df=new_df.sort_values(by=['production'],ascending=False)
new_df
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=new_df,x="state",y="production")
plt.xticks(rotation=90)
plt.show()
```

```python
x=new_df.head()
x
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=x,x="state",y="production")
plt.xticks(rotation=90)
plt.show()
```

```python
#northdakota,california,southdakota,florida,montana are top 5 honey producing state
```

# 3.What is the Change in mean Average price of Honey

```python
df.groupby('year')["average_price"].mean()
```

```python
new_df=df.groupby("year")["average_price"].mean().reset_index()
```

```python
new_df
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=new_df,x='year',y="average_price")
plt.show()
plt.figure(figsize=(15,4))
sns.barplot(data=new_df,x='year',y="average_price")
plt.show()
```

```python
#1995-2017 the avg price of honey is inc in a good manner but we ca observer a udden drop in the avg price after 2017 and after
#it is constant
```

# Which was the year when production of Honey in whole US was the highest?

```python
df.groupby('year')['production'].sum()
```

```python
highest_production=df.groupby('year')['production'].sum()
```

```python
highest_production
```

```python
highest_production=highest_production.sort_values(ascending=False).reset_index()
```

```python

plt.figure(figsize=(10,5))
sns.barplot(x=highest_production.year,y=highest_production.production)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#the highest production is in 2000 year
```

# Q5.from the above inference we get the production was highest in the year 2000, now let infer which state was having highest contribution in that year?

```python
max_state_2000 = df[df['year']==2000]
max_state_2000
```

```python
max_state_2000=max_state_2000.sort_values('production',ascending=False)
max_state_2000
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=max_state_2000,x='state',y='production')
plt.xticks(rotation=90)
plt.show()
```
